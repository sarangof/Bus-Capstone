```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
import seaborn as sns; sns.set()
```

```python
datacb = pd.read_csv('combined_trip_summary.csv')
date = datacb['schedule_date']
datacb = datacb[['schedule_date', 'aggregate_trip_time']]
date = map(lambda x: x[0:10], date)
datacb['date'] = date
Month = datacb['schedule_date']
Month = map(lambda x: x[5:7], Month)
datacb['Month'] = Month
datacb.head()
```

```python
datacb_date = datacb.groupby('date').sum()
datacb_date.head()
```

```python
datacb_date.to_csv('datacb_date.csv')
```

```python
datacb_date = pd.read_csv('datacb_date.csv')
datacb_date
```

```python
# datacb = pd.read_csv('combined_trip_summary.csv')
# datacb = datacb[['schedule_date', 'aggregate_trip_time']]
Month = datacb_date['date']
Month = map(lambda x: x[5:7], Month)
datacb_date['Month'] = Month
record = datacb_date['aggregate_trip_time']
record = map(lambda x: x/30, record)
datacb_date['record'] = record
datacb_date['date'] = pd.to_datetime(datacb_date.date)
datacb_date.head()
```

```python
data_count = pd.read_csv('count_date.csv')
data_count = data_count.iloc[1:]
Month = data_count['Date_new']
Month = map(lambda x: x[5:7], Month)
data_count['Month'] = Month
data_count['Date_new'] = pd.to_datetime(data_count.Date_new)
data_count.head()
```

```python
data_count.columns = ['date', 'Count', 'Month']
```

```python
data_merge = datacb_date.merge(data_count, on='date', how='outer')
```

```python
data_merge.head()
```

```python
data_meage = data_merge[['date', 'aggregate_trip_time', 'Month_x', 'record', 'Count']]
data_meage.columns = ['date', 'aggregate_trip_time', 'Month', 'record', 'Count']
data_meage.head()
```

```python
data_meage_1 = data_meage[data_meage['Month'] == '08'].dropna()
data_meage_1
```

```python
np.corrcoef(data_meage_1['record'],data_meage_1['Count'])[0][1]
```

```python
for i in ['01','02','03','04','05','06','07','08','09','10','11','12']:
    data_meage_1 = data_meage[data_meage['Month'] == i].dropna()
    corr = np.corrcoef(data_meage_1['record'],data_meage_1['Count'])[0][1]
    print 'For month {}, correlation coefficient is {}'.format(i, corr)
```
